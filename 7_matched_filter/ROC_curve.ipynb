```python
from sklearn import metrics
import numpy as np
from matplotlib import pyplot as plt
```

```python
def concat(arr1,arr2):
    return np.concatenate((arr1,arr2))

def fprtpr(truearr, predarr):
    fpr,tpr,_ = metrics.roc_curve(truearr, predarr)
    return fpr, tpr
```

```python
# Load eta 160
eta160g1sig = np.load("eta160_g1_sigpbkg.npy")
eta160g1bkg = np.load("eta160_g1_bkg.npy")

eta160g2sig = np.load("eta160_g2_sigpbkg.npy")
eta160g2bkg = np.load("eta160_g2_bkg.npy")

eta160g3sig = np.load("eta160_g3_sigpbkg.npy")
eta160g3bkg = np.load("eta160_g3_bkg.npy")

# Load eta 100
eta100g1sig = np.load("eta100_g1_sigpbkg.npy")
eta100g1bkg = np.load("eta100_g1_bkg.npy")

eta100g2sig = np.load("eta100_g2_sigpbkg.npy")
eta100g2bkg = np.load("eta100_g2_bkg.npy")

eta100g3sig = np.load("eta100_g3_sigpbkg.npy")
eta100g3bkg = np.load("eta100_g3_bkg.npy")

# Load eta 50
eta50g1sig = np.load("eta50_g1_sigpbkg.npy")
eta50g1bkg = np.load("eta50_g1_bkg.npy")

eta50g2sig = np.load("eta50_g2_sigpbkg.npy")
eta50g2bkg = np.load("eta50_g2_bkg.npy")

eta50g3sig = np.load("eta50_g3_sigpbkg.npy")
eta50g3bkg = np.load("eta50_g3_bkg.npy")

# True vals
sigtrue = np.ones(500)
bkgtrue = np.zeros(500)
```

```python
truearr = np.concatenate((sigtrue,bkgtrue))
```

```python
eta160g1fpr, eta160g1tpr = fprtpr(truearr,concat(eta160g1sig,eta160g1bkg))
eta160g2fpr, eta160g2tpr = fprtpr(truearr,concat(eta160g2sig,eta160g2bkg))
eta160g3fpr, eta160g3tpr = fprtpr(truearr,concat(eta160g3sig,eta160g3bkg))

eta100g1fpr, eta100g1tpr = fprtpr(truearr,concat(eta100g1sig,eta100g1bkg))
eta100g2fpr, eta100g2tpr = fprtpr(truearr,concat(eta100g2sig,eta100g2bkg))
eta100g3fpr, eta100g3tpr = fprtpr(truearr,concat(eta100g3sig,eta100g3bkg))

eta50g1fpr, eta50g1tpr = fprtpr(truearr,concat(eta50g1sig,eta50g1bkg))
eta50g2fpr, eta50g2tpr = fprtpr(truearr,concat(eta50g2sig,eta50g2bkg))
eta50g3fpr, eta50g3tpr = fprtpr(truearr,concat(eta50g3sig,eta50g3bkg))
```

```python
TPRvFPR, ax1 = plt.subplots()

lw=2

ax1.plot(eta160g1fpr, eta160g1tpr,label='g = 1',color='darkorange',lw=lw)
ax1.plot(eta160g2fpr, eta160g2tpr,label='g = 2',color='red',lw=lw)
ax1.plot(eta160g3fpr, eta160g3tpr,label='g = 3',color='maroon',lw=lw,linestyle='--')

ax1.plot([0, 1], [0, 1], color='grey', lw=lw, linestyle='--')
ax1.set_xlim([-0.01, 1.0])
ax1.set_ylim([0.0, 1.05])
ax1.set_xlabel('False Positive Rate')
ax1.set_ylabel('True Positive Rate')
ax1.legend(loc='lower right')

ax1.set_title(r'$\eta = 160$ Mpc')
#TPRvFPR.savefig('eta160_MF_result.pdf')
```
